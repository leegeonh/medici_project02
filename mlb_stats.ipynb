```python
statsapi.get('league', {'sportId': 1, 'leagueIds':103})['leagues'][0].keys()
```

```python
leagues = []
for id in league_ids: 
    leagues.append(statsapi.get('league', {'sportId': 1, 'leagueIds':id})['leagues'][0])
    time.sleep(1)
```

```python
league_cols = ['id', 'name', 'link', 'abbreviation', 'nameShort', 'seasonState', 'hasWildCard', 'hasSplitSeason', 'numGames', 'hasPlayoffPoints',
               'numTeams', 'numWildcardTeams', 'season', 'orgCode', 'conferencesInUse', 'divisionsInUse', 'sortOrder', 'active',
               'seasonDateInfo.seasonId', 'seasonDateInfo.regularSeasonStartDate', 'seasonDateInfo.regularSeasonEndDate', 
               'seasonDateInfo.lastDate1stHalf', 'seasonDateInfo.allStarDate', 'seasonDateInfo.firstDate2ndHalf',  
               'seasonDateInfo.postSeasonStartDate', 'seasonDateInfo.postSeasonEndDate',
               'seasonDateInfo.qualifierPlateAppearances', 'seasonDateInfo.qualifierOutsPitched']

df_leagues = pd.json_normalize(leagues)[league_cols]
df_leagues
```
